# 알고리즘 시작~

```python
#사용자 선택 맥주 클러스터 번호 가져오기
cluster_num=[]
for beers in choose:
    for cl, beer in options.items():
        for i in range(len(beer)):
            if beers in beer[i]:
                cluster_num.append(cl)
```

## 음식 필터링

```python
#음식 필터링
if '음식' in which_cat:
    if food=='빵류': get_file=bread
    elif food=='치즈류': get_file=cheese
    elif food=='마른안주': get_file=dry
    elif food=='과일': get_file=fruit
    elif food=='지중해': get_file=med
    elif food=='견과류': get_file=nuts
    elif food=='풍미 있는 음식(치즈, 훈제요리 등)': get_file=rich
    elif food=='해산물': get_file=seafood
    elif food=='매운 향의 음식': get_file=spicy
    else:
        get_file=sweets

beer_index_filtered=get_file.loc[get_file['Quantile']==str(adventure)+'Q'].index.values
```

```python
#주.의. 이름으로 필터링 하기 (여기 인덱스랑 저기 인덱스 안 맞음)
name_filtered1=get_file.iloc[beer_index_filtered,1]
name_filtered1
```

```python
#이거 수정! 처음부터 wrong_cluster 만들지 말고 beer_index_filtered에서 바로 빠지는 방법으로다가

beer_index_filtered2=[] #지우는 방식 말고 필요한 것들만 저장

for idx in beer_index_filtered:
    if get_file['cluster'][idx] in cluster_num:
        beer_index_filtered2.append(idx)
```

```python
#필터링 한번 더 ~ 클러스터링 해당 번호에 맞춰서 -> beer_index_filtered로 하고 나중에 beer_name_filtered 까지 뽑기

beer_name_filtered=list(get_file.iloc[beer_index_filtered2,1])
```

```python
beer_name_filtered  #여기까지 오케이
```

```python
#피어슨용 인덱스 찾아주기
pearson_index=[]
for beer in beer_name_filtered:
    for i in range(pearson.shape[0]):
        if beer==pearson.iloc[i,1]:
            pearson_index.append(i)
print(len(pearson_index)==len(beer_name_filtered))
```

```python
#피어슨 유사도 대입 ~ 튜플 쌍으로 받기 (맥주 이름, 유사도, 클러스터) 이런 식으로 
#주.의. 인덱스 접근 말고 (할 거면 피어슨 유사도에서 새로 받기) 이름 접근하기

candidate=[]
for beer in choose:
    for name,idx in zip(beer_name_filtered,pearson_index):
        if beer!=name:
            put=(name,pearson[pearson['name_kor_x']==beer][name].values,pearson.iloc[idx,-1],beer)
            candidate.append(put)
```

```python
semi_get=sorted(candidate, key=lambda x: x[1],reverse=True)
```

```python
semi_get
```

```python
##클러스터 개수에 대한 결과 -> 일단 10개라 생각하고 뽑아, 10개보다 적은 경우에는 끝에 x개 탈락시키자

recommendation_list=[]
for i in range(len(semi_get)):
    if len(set(cluster_num))==3:
        for a in range(4):
            if semi_get[i][2]==sorted(set(cluster_num),key=cluster_num.index)[0]:
                recommendation_list.append([semi_get[i][0],semi_get[i][2]])
        for b in range(3):
            if semi_get[i][2]==sorted(set(cluster_num),key=cluster_num.index)[1]:
                recommendation_list.append([semi_get[i][0],semi_get[i][2]])
        for c in range(3):
            if semi_get[i][2]==sorted(set(cluster_num),key=cluster_num.index)[2]:
                recommendation_list.append([semi_get[i][0],semi_get[i][2]])
    elif len(set(cluster_num))==2:
        for a in range(7):
            if semi_get[i][2]==sorted(set(cluster_num),key=cluster_num.index)[0]:
                recommendation_list.append([semi_get[i][0],semi_get[i][2]])
        for b in range(3):
            if semi_get[i][2]==sorted(set(cluster_num),key=cluster_num.index)[1]:
                recommendation_list.append([semi_get[i][0],semi_get[i][2]])
    else:
        for a in range(10):
            if semi_get[i][2]==sorted(set(cluster_num),key=cluster_num.index)[0]:
                recommendation_list.append([semi_get[i][0],semi_get[i][2]])                    
```

```python
recommendation_list
```

```python
np.unique(recommendation_list).tolist()
```

```python
final_recom=np.unique(recommendation_list).tolist()[len(set(cluster_num)):10+len(set(cluster_num))] #낫배드!
print(final_recom)
```
