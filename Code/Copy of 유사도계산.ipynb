```python
num_feature = ['apv', 'ibu', 'score_final', 'First_sit_score', 'Second_sit_score', 'Third_sit_score', 'V3', 'V4', 'V5','V6','V7', 'V8', 'V9', 'V10','V11','V12','V13']
df_num = df[num_feature]
```

```python
df_num_apv = df_num.dropna(subset=['apv'], how='all')
df_num_final = df_num.dropna(subset=['score_final'], how='all')
df_num_first = df_num.dropna(subset=['First_sit_score'], how='all')
df_num_second = df_num.dropna(subset=['Second_sit_score'], how='all')
df_num_third = df_num.dropna(subset=['Third_sit_score'], how='all')
```

```python
df_num.shape
```

```python
idx=[102,201,341,342,398,405,443,653,680,734,746,770,771]; values=[0,10,70,70,40,85,10,0,85,70,0,10,10]

for i in range(len(idx)):
  df_num.iloc[idx[i],1]=values[i]
  #df_num['ibu'].loc[i,]=values[i]
#df_num['ibu'].loc[102,201,341,342,398,405,443,653,680,734,746,770,771] =  [0,10,70,70,40,85,10,0,85,70,0,10,10]
```

```python
df_num['apv'].fillna(df_num_apv['apv'].mean(), inplace=True)
df_num['score_final'].fillna(df_num_final['score_final'].mean(), inplace=True)
df_num['First_sit_score'].fillna(df_num_first['First_sit_score'].mean(), inplace=True)
df_num['Second_sit_score'].fillna(df_num_second['Second_sit_score'].mean(), inplace=True)
df_num['Third_sit_score'].fillna(df_num_third['Third_sit_score'].mean(), inplace=True)
```
