```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome('/usr/bin/chromedriver', options=options)
```

```python
url_base="https://www.wine21.com/13_search/beer_view.html?Idx="
```

```python
#exception 처리 대비
from selenium.common.exceptions import UnexpectedAlertPresentException, NoSuchElementException
```

```python
#첫 페이지만 가져와보기

import time

name_kor=[]; name_eng=[]; producer=[]; region=[]; style=[]; apv=[]

for num in range(500999,500800,-1):
    url=url_base+str(num)
    driver.get(url)
    time.sleep(2)
    print("currently on: ",num, end=" ")
    #name_kor
    try:
        name_kor.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[1]/h4').text)
    except UnexpectedAlertPresentException:
        pass
    except NoSuchElementException:
        pass

    #name_eng
    try:
        name_eng.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[1]/div[1]').text)
    except UnexpectedAlertPresentException:
        pass
    except NoSuchElementException:
        pass

    #producer
    try:
        producer.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[1]').text)
    except UnexpectedAlertPresentException:
        pass  
    except NoSuchElementException:
        pass

    #region
    try:
        region.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[2]').text)
    except UnexpectedAlertPresentException:
        pass 
    except NoSuchElementException:
        pass

    #style
    try:
        style.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[3]').text)
    except UnexpectedAlertPresentException:
        pass 
    except NoSuchElementException:
        pass
    
    #apv
    try:
        apv.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[4]').text)
    except UnexpectedAlertPresentException:
        pass 
    except NoSuchElementException:
        pass
        
    #name_kor.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[1]/h4'))
    #name_eng.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[1]/div[1]'))
    #producer.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[1]'))
    #region.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[2]'))
    #style.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[3]'))
    #apv.append(driver.find_element_by_xpath('//*[@id="frmMain"]/article/div/div[2]/div[5]/dl/dd[4]'))
```
